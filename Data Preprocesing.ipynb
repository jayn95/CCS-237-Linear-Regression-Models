```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import LabelEncoder

# --- Load dataset ---
df = pd.read_csv('data\imdb_top_1000.csv')

# --- Clean Runtime ("142 min" â†’ 142) ---
df['Runtime'] = df['Runtime'].replace(' min', '', regex=True)
df['Runtime'] = pd.to_numeric(df['Runtime'], errors='coerce')

# --- Clean Gross ("$123.4M" â†’ 123.4 * 1e6) ---
def clean_gross(val):
    if pd.isna(val):
        return np.nan
    val = str(val).replace('$', '').replace(',', '').strip()
    try:
        if 'M' in val:
            return float(val.replace('M', '')) * 1e6
        elif 'B' in val:
            return float(val.replace('B', '')) * 1e9
        else:
            return float(val)
    except:
        return np.nan

df['Gross'] = df['Gross'].apply(clean_gross)

# --- Convert Meta_score to numeric ---
df['Meta_score'] = pd.to_numeric(df['Meta_score'], errors='coerce')

# --- Encode Genre ---
if df['Genre'].dtype == 'object':
    le = LabelEncoder()
    df['Genre'] = le.fit_transform(df['Genre'])

# --- Select relevant columns ---
cols = ['Released_Year', 'Runtime', 'Genre', 'Meta_score', 'No_of_Votes', 'Gross', 'IMDB_Rating']
df = df[cols]

# --- Drop or fill missing data ---
df = df.dropna()  # Drop rows with any missing values

# --- Remove infinite values just in case ---
df = df.replace([np.inf, -np.inf], np.nan).dropna()

# --- Define independent and dependent variables ---
X = df[['Released_Year', 'Runtime', 'Genre', 'Meta_score', 'No_of_Votes', 'Gross']]
y = df['IMDB_Rating']

# Ensure numeric
X = X.apply(pd.to_numeric, errors='coerce')

# Drop again if any nans slipped in
X = X.replace([np.inf, -np.inf], np.nan).dropna()

# --- Add constant for VIF ---
X_const = sm.add_constant(X)

# --- Compute VIF ---
vif_data = pd.DataFrame()
vif_data["Variable"] = X_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]

print("\nâœ… Variance Inflation Factor (VIF) Results:")
print(vif_data.round(2))

# --- Optional: Correlation matrix ---
print("\nðŸ“Š Correlation Matrix:")
print(X.corr().round(2))
```
